```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('AccidentesAviones.csv', index_col=0)
df.head(10)
```

```python
#observamos los tipos de datos
df.info()
```

```python
#eliminamos columnas que no necesitaremos
df2= df.drop(columns=['flight_no', 'route', 'registration','cn_ln'])
```

```python
df2.head()
```

```python
#reemplazamos simbolos
df2.replace("?",np.nan, inplace=True)
```

```python
#cambiamos el formato de fechas
df2[['mes','dia','año']] = df2['fecha'].str.split(' ', expand=True)
df2['dia'] = df2.dia.str.replace(',', '')
df2['mes'] = pd.to_datetime(df2['mes'], format= '%B').dt.month
df2['fecha'] = df2['año'].astype(str)+'-'+df2['mes'].astype(str)+'-'+df2['dia'].astype(str)
df2.head()
```

```python
df2.drop(columns='HORA declarada', inplace=True)
```

```python
#cambiamos nombres de columnas
df2.rename(columns={'OperadOR': 'operador', 'PASAJEROS A BORDO': 'pasajeros_a_bordo', 'crew_aboard': 'personal_a_bordo', 'all_aboard': 'total_a_bordo',
           'passenger_fatalities': 'pasajeros_fallecidos', 'crew_fatalities': 'personal_fallecido', 'cantidad de fallecidos': 'total_fallecidos',
           'ac_type': 'tipo_de_avion','Ruta': 'ruta'}, inplace=True)
df2.head()
```

```python
df2= pd.DataFrame(df2)
df2.to_csv('aviones_arreglado.csv', index= False)
```
